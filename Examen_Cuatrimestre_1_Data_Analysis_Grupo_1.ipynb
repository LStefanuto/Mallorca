## Limpieza de datos
### Subset de la Data para comenzar el análisis
Primero, evaluamos los mercados disponibles en el archivo y la cantidad de registros para cada uno. 
Al observar que los registros del mercado de **Mallorca** supera ampliamente al resto (16655 registros en total) seleccionamos este para análizar, desestimando asi "Barcelona" y "Other (International).

Al nuevo subset seleccionado lo denominamos **mallorca**.

```
mercados=listado['market'].value_counts().reset_index(name='total')
print(mercados)
sns.barplot(y='index', x='total', data=mercados, orient='h')

f = pd.Series(['Mallorca Island','Mallorca Other','Palma'])
mallorca = listado[listado['market'].isin(f)]
```

###Data Cleaning de "Mallorca Island" subset
Comenzamos por borrar las **columnas y filas obsoletas**, cuyos registros eran todos nulos y aquellas columnas vacías o cuyos valores eran todos "none".

```
# Borramos columnas obsoletas
mallorca.dropna(axis=1, how='all')

# Borramos filas obsoletas
mallorca.dropna(how='all')

# Borramos columnas con none, o sin datos
del mallorca['experiences_offered']
del mallorca['xl_picture_url']
del mallorca['host_acceptance_rate']
del mallorca['neighbourhood']
del mallorca['neighbourhood_group_cleansed']
del mallorca['jurisdiction_names']
del mallorca['thumbnail_url']
del mallorca['medium_url']
```

Luego, evaluamos la presencia de nulos y duplicados: 

*   Los **nulos** en datos numéricos fueron reemplazados por 0. En cambio, los de datos de tipo string, fueron reemplazados por un espacio " ".
*   Las **filas duplicadas** fueron eliminadas ya que corresponderían a una misma review (si son distintas el host id y listing is es distinto).


```
# Verificamos nulos
len(mallorca.isnull()) 

# Verificamos si hay duplicados
mallorca.duplicated()
```

```
# Completamos nulos con 0 y borramos duplicados
mallorca.drop_duplicates()
reviews.drop_duplicates()

# Asi podemos llenar los nulos dependiendo del tipo de dato
float_cols = mallorca.select_dtypes(include=['float64']).columns
str_cols = mallorca.select_dtypes(include=['object']).columns

mallorca.loc[:, float_cols] = mallorca.loc[:, float_cols].fillna(0)
mallorca.loc[:, str_cols] = mallorca.loc[:, str_cols].fillna('')
```

###Data Cleaning del archivo "Reviews"
Evaluamos la presencia de nulos y duplicados y eliminamos en el caso que hubiese una fila entera con nulos o duplicada.

Los nulos y duplicados fueron reemplazados igual que como se hizo en el subset "mallorca", teniendo en cuenta el tipo de dato.

```
reviews.drop_duplicates()

float_cols = reviews.select_dtypes(include=['float64']).columns
str_cols = reviews.select_dtypes(include=['object']).columns

reviews.loc[:, float_cols] = reviews.loc[:, float_cols].fillna(0)
reviews.loc[:, str_cols] = reviews.loc[:, str_cols].fillna('')
```

## Transformación de datos

Comenzamos por quitas el signo $ del valor de la columna "price" para poder así, considerarlo como un número. Transformamos este valor a un dato tipo float para poder trabajar luego con esta variable.

```
# Quitamos signo pesos y nos aseguramos que se lo considere como un número
mallorca['price'] = mallorca['price'].str.replace('$','').str.replace(',','')
mallorca['price'] = mallorca['price'].astype('float')
```

Los dos primeros gráficos muestran la cantidad de comentarios en Airbnb. En el primero diariamente y en el segundo el acumulado mensual.
